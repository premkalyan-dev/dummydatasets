```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel("Q11_website_traffic_dummy_dataset.xlsx")

df["Date"] = pd.to_datetime(df["Date"])
daily_traffic = df.groupby("Date")["Visits"].sum().reset_index()

plt.figure(figsize=(12, 5))
sns.lineplot(data=daily_traffic, x="Date", y="Visits", color="blue")
plt.title("Daily Website Traffic")
plt.xlabel("Date")
plt.ylabel("Total Visits")
plt.tight_layout()
plt.show()

traffic_by_source = df.groupby("Traffic Source")["Visits"].sum().reset_index().sort_values(by="Visits", ascending=False)

plt.figure(figsize=(8, 5))
sns.barplot(data=traffic_by_source, x="Traffic Source", y="Visits", palette="Set2")
plt.title("Traffic by Source")
plt.tight_layout()
plt.show()

traffic_by_page = df.groupby("Page URL")["Visits"].sum().reset_index().sort_values(by="Visits", ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(data=traffic_by_page, x="Page URL", y="Visits", palette="Set3")
plt.title("Top 10 Visited Pages")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
