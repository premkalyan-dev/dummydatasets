```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel(r"Q13_revenue_drop_analysis_dataset.xlsx")
df["Date"] = pd.to_datetime(df["Date"])
df["Quarter"] = df["Date"].dt.to_period("Q")

revenue_by_quarter = df.groupby("Quarter")["Order Value"].sum().reset_index()
volume_by_quarter = df.groupby("Quarter")["Units Sold"].sum().reset_index()
aov_by_quarter = df.groupby("Quarter").apply(lambda x: x["Order Value"].sum() / x.shape[0]).reset_index(name="Average Order Value")
retention_by_quarter = df.groupby("Quarter")["Is Repeat Customer"].mean().reset_index()
retention_by_quarter["Retention %"] = retention_by_quarter["Is Repeat Customer"] * 100

plt.figure(figsize=(10, 5))
sns.barplot(data=revenue_by_quarter, x="Quarter", y="Order Value", palette="Blues_d")
plt.title("Total Revenue by Quarter")
plt.ylabel("Revenue ()")
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 5))
sns.barplot(data=volume_by_quarter, x="Quarter", y="Units Sold", palette="Greens_d")
plt.title("Total Units Sold by Quarter")
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 5))
sns.barplot(data=aov_by_quarter, x="Quarter", y="Average Order Value", palette="Oranges_d")
plt.title("Average Order Value by Quarter")
plt.ylabel("")
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 5))
sns.barplot(data=retention_by_quarter, x="Quarter", y="Retention %", palette="Purples_d")
plt.title("Customer Retention Rate by Quarter")
plt.ylabel("Repeat Customer %")
plt.tight_layout()
plt.show()
```
